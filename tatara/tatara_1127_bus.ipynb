```python
import numpy as np
import numpy.random as random
import scipy as sp
import pandas as pd
from pandas import Series,DataFrame

# 可視化モジュール
import matplotlib.pyplot as plt
import matplotlib as mpl
#import seaborn as sns
%matplotlib inline

# 小数第３まで表示
%precision 3
```

```python
Collision_data = pd.read_csv('NYC_accidents_heat_map.csv')
Collision_data.head()
```

```python
Collision_data['la'] = Collision_data['LATITUDE'].map(lambda x : round(float(x),3))
Collision_data['lo'] = Collision_data['LONGITUDE'].map(lambda x : round(float(x),3))
```

```python
bus_data = Collision_data[Collision_data['VEHICLE TYPE CODE 1']=='BUS']
```

```python
df = bus_data.loc[:,['CONTRIBUTING FACTOR VEHICLE 1','la','lo']]
```

```python
df = df.dropna()
df.info()
```

```python
Dummy = pd.get_dummies(df['CONTRIBUTING FACTOR VEHICLE 1'])
Dummy.info()
```

```python
a1 = pd.concat([df.loc[:,['la','lo']], Dummy], axis=1)
a1.head()
```

```python
df2 = a1.groupby(['la','lo']).sum()
```

```python
a1.sum()
```

```python
#事故が少ない要因は減らす

df3 = df2
for i in range(len(df2.columns)):
    if df2.iloc[:, len(df2.columns) - i -1].sum()<100 :
        df3 = df3.drop(a1.columns[len(a1.columns) - i -1],axis=1)
```

```python
df3.head()
```

```python
df3 = df3.drop('Unspecified', axis = 1)
```

```python
df3.head()
```

```python
a=[]
i=0
for i in range(len(df3.index)):
    a.append(df3.iloc[i,:].sum())
af = pd.DataFrame(a,index = df3.index,columns={'sum'})
```

```python
af.head()
```

```python
af.info()
```

```python
df3.info()
```

```python
df4 = pd.concat([df3,af],axis=1)
```

```python
df4.head()
```

```python
df5 = df4.sort_values(by='sum', ascending=False)
```

```python
df5 = df5.head(20)
df5
```
